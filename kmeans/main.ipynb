```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# create dataset
data = []
for i in range(50):
    point = [random.uniform(0, 5), random.uniform(0,4)]
    data.append(point)

for i in range(50):
    point = [random.uniform(3, 10), random.uniform(3,10)]
    data.append(point)

k1_x = random.uniform(0, 10)
k1_y = random.uniform(0, 10)
dist1 = 0
ss_list = []
k2_x = random.uniform(0, 10)
k2_y = random.uniform(0, 10)
dist2 = 0


#training
def python_list_add(in1, in2):
    wrk = np.array(in1) + np.array(in2)
    return wrk.tolist()

for i in range(100):

    total_dist =0
    num_k1 = 1
    num_k2 = 1
    next_k1 = [0,0]
    next_k2 = [0,0]
    for point in data:
        dist1 = (point[0] - k1_x)*(point[0] - k1_x) + (point[1] - k1_y)*(point[1] - k1_y)
        dist2 = (point[0] - k2_x)*(point[0] - k2_x) + (point[1] - k2_y)*(point[1] - k2_y)
        if dist1 <= dist2:
            next_k1 = python_list_add(next_k1, point)
            total_dist += dist1
            num_k1 +=1
        else:
            next_k2 = python_list_add(next_k2, point)
            total_dist += dist2
            num_k2 +=1
        # renew 2 points
    k1_x = next_k1[0] / num_k1
    k1_y = next_k1[1] / num_k1
    k2_x = next_k2[0] / num_k2
    k2_y = next_k2[1] / num_k2
    ss_list.append(total_dist)


plot_x = []
plot_y = []
colourlist = []
for point in data:
    dist1 = (point[0] - k1_x)*(point[0] - k1_x) + (point[1] - k1_y)*(point[1] - k1_y)
    dist2 = (point[0] - k2_x)*(point[0] - k2_x) + (point[1] - k2_y)*(point[1] - k2_y)
    if dist1 <= dist2:
        plot_x.append(point[0])
        plot_y.append(point[1])
        colourlist.append("c")
    else:
        plot_x.append(point[0])
        plot_y.append(point[1])
        colourlist.append("m")

plt.scatter(plot_x, plot_y, color=colourlist)
```
