```python
import random
import cupy as np
import matplotlib.pyplot as plt
import time


n = 10000000
m = 10000000
p = np.pi


score = 0 
sc_list = []
r_ans = 3.14159265359

start =time.time()
for i in range(m):
  x = np.random.rand(n, 1)
  y = np.random.rand(n, 1)
  r_sq = np.square(x) + np.square(y)
  score += np.count_nonzero(r_sq <= 1)
  sc_list.append(score / n / (i +1) * 4)
  if (i+1) % 1000 ==0:
    print((i+1)*n, ",", (i+1) / m * 100, "%")

end = time.time()
print("pi:", score / n / m * 4)
print("pi:", p)
print("accuracy:", score/ n / m * 4 / p, "%")
print("time:", (end - start) / 60)
fig = plt.plot(sc_list)
plt.show()
```
