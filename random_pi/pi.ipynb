```python
import random
import cupy as np
import matplotlib.pyplot as plt


n = 100000000
m = 100000
score = 0 
sc_list = []
for i in range(m):
  x = np.random.rand(n, 1)
  y = np.random.rand(n, 1)
  r_sq = np.square(x) + np.square(y)
  score += np.count_nonzero(r_sq <= 1)
  sc_list.append(score / n / (i +1) * 4)

print(score / n / m * 4)
fig = plt.plot(sc_list)
plt.show()
```
